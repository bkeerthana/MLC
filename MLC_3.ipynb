Lifecycle of DataAnalytics

     Problem Definition
     Data Collection
     Data Cleaning
     Exploratory Data Analysis
     Data Modeling
     Data Visualization
     Decision Making


| Step                                | Focus (What this step concentrates on)       | Why this step is important                    | Key Keywords                                | Simple Example                                                           |
| ----------------------------------- | -------------------------------------------- | --------------------------------------------- | ------------------------------------------- | ------------------------------------------------------------------------ |
| **Problem Definition**              | Clearly defining the question to be answered | Prevents unclear or irrelevant analysis       | Objective, scope, stakeholders, constraints | “Identify phishing emails to reduce users clicking malicious links.”     |
| **Data Collection**                 | Gathering relevant and sufficient data       | Analytics quality depends on data quality     | Data sources, logs, fields, attributes      | Collect sender domain, subject, links, attachment type, user action.     |
| **Data Cleaning**                   | Improving data quality by fixing errors      | Dirty data leads to wrong conclusions         | Missing values, duplicates, normalization   | Remove duplicate emails and fix inconsistent domain names.               |
| **Exploratory Data Analysis (EDA)** | Understanding patterns and behavior in data  | Helps discover trends and anomalies           | Distribution, frequency, outliers, trends   | Phishing emails contain more links and urgent words.                     |
| **Data Modeling**                   | Creating logic or models to detect phishing  | Enables prediction or classification          | Features, labels, rules, algorithms         | Classify emails as phishing or legitimate using link count and keywords. |
| **Data Visualization**              | Communicating insights clearly               | Helps stakeholders understand results quickly | Charts, graphs, dashboards                  | Bar chart showing phishing vs legitimate emails.                         |
| **Decision Making**                 | Taking action based on analytics insights    | Converts analysis into real-world impact      | Risk mitigation, policy, action             | Block emails with high phishing probability and train users.             |

Data Collection:

    Focus:
    Identifying what data is required and where it comes from.

    Why this step matters:
    Wrong or insufficient data leads to wrong conclusions.

    Example:
    Activity: Identify data requirements


Activity:

    What fields are needed to detect phishing
    
    Which data should NOT be collected (privacy)

Data Cleaning:


    Focus:
    ensures the collected data is accurate, consistent, and usable.

    why matters:
    Dirty data produces incorrect analytics results.

    Activity: Spot the Problem

Exploratory Data Analysis (EDA):


    Focus:
    This step focuses on understanding patterns and behavior in the data.
    
    Why matters:
    EDA reveals insights before building models.
    
    Ask to write:
    Two observations
    One question raised by the data



Data Modeling:

    This step focuses on creating logic or models to classify or predict outcomes.

    Models automate decision-making.

    Ask to design simple rules:
    “If email has more than X links and urgent words → phishing”


Data Visualization:

    Focus:
    This step focuses on communicating insights clearly to others.
    
    why it matters:
    Insights that cannot be explained are not useful.

    Activity:
    Chart selection

Decision Making:


    This step focuses on converting insights into real actions.
