```python
# 1) Read SQL file (or paste your SQL content as a string)
with open(datafile, "r", encoding="utf-8") as f:
    sql = f.read()
```

```python
import re
```

```python
# 2) Remove MySQL-specific statements
#re.sub(r"^\s*--.*$", "", sql, flags=re.IGNORECASE | re.MULTILINE)
sql = re.sub(r"(?im)^\s*--.*$", "", sql)                 # remove comment lines starting with --
sql = re.sub(r"(?im)^\s*CREATE DATABASE.*?;\s*$", "", sql)
sql = re.sub(r"(?im)^\s*USE\s+\w+\s*;\s*$", "", sql)
```

```python
# 3) Convert MySQL AUTO_INCREMENT to SQLite AUTOINCREMENT
#    and make INT PRIMARY KEY AUTO_INCREMENT -> INTEGER PRIMARY KEY AUTOINCREMENT
sql = re.sub(
    r"(?i)\blog_id\s+INT\s+PRIMARY\s+KEY\s+AUTO_INCREMENT\b",
    "log_id INTEGER PRIMARY KEY AUTOINCREMENT",
    sql
)
```

```python
# 4) Create SQLite DB in memory and execute script
conn = sqlite3.connect(":memory:")
conn.executescript(sql)

# 5) Read table into pandas DataFrame
df = pd.read_sql("SELECT * FROM auth_logs;", conn)

df
```
