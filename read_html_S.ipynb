```python
import pandas as pd

base = "https://www.fdic.gov/bank-failures/failed-bank-list"
all_pages = []

for pg in range(0, 50):  # 50 is a safe upper bound; we will stop when no table / no rows
    url = f"{base}?items_per_page=25&pg={pg}"
    try:
        tables = pd.read_html(url)
        if not tables:
            break
        page_df = tables[0]
        if page_df.empty:
            break
        all_pages.append(page_df)
    except ValueError:
        # read_html raises ValueError when it can't find tables
        break

df = pd.concat(all_pages, ignore_index=True).drop_duplicates()
print(df.shape)
df.head()
```
